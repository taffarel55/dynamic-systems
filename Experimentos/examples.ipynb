## 2. To study transient response for given system

```octave
s=%s;

z=0.9;
num=1;den=36+2*z*s+s^2;
TF = syslin('c', num, den);
subplot(131);
t=0:0.1:10;
y1 = csim('step', t, TF);
title('Subamortecido '); plot(t, y1);

z=6;
den =36+2*z*s+s^2;
TF = syslin('c', num, den);
subplot(132);
y1 = csim('step', t, TF);
title('Criticamente amortecido'); 
plot(t, y1);

z=20;
den = 36+2*z*s+s^2;
TF = syslin('c', num, den);
subplot(133);
t=0:0.1:10;
y1=csim('step', t, TF);
title('Superamortecido'); plot(t, y1)
```

## 4. To compare open loop and closed loop system.

```octave
s = %s;
z = 0.1;
num = 1;den=36+2*z*s+s^2;
TF = syslin('c', num, den);
disp(TF);
subplot(121);
t=0:0.1:30;
y1 = csim('step', t, TF);
title('Malha aberta');
plot(t, y1);

num1=s; den1=1+s;
TF1 = syslin('c', num1, den1);
num2 =1; den2 =3;
TF2 = syslin('c', num2, den2);
S1=TF1/.TF2;
disp(S1);
subplot(122);
t=0:0.1:30;
y1 = csim('step', t, S1);
title('Malha fechada');
plot(t, y1)
```
